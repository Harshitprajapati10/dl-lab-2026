```python
import torch
```

```python
marks = torch.zeros(80, 5, dtype=torch.long)
marks
```

```python
choice = torch.randint(0, 3, (80,5))
marks[choice == 0] = torch.randint(1, 41, ((choice == 0).sum(),))
marks[choice == 1] = torch.randint(41, 71, ((choice == 1).sum(),))
marks[choice == 2] = torch.randint(71, 96, ((choice == 2).sum(),))
```

```python
marks
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
marks_np = marks.numpy()
plt.figure(figsize=(8, 12))
sns.heatmap(marks_np, cmap="viridis", cbar=True)
plt.title("Heatmap of Marks (80 Students  5 Subjects)")
plt.xlabel("Subjects")
plt.ylabel("Students")
plt.show()
```
