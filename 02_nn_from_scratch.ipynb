## Generate the dataset

```python
import torch
```

```python
marks = torch.zeros(80, 6, dtype=torch.long)
marks
```

```python
choice = torch.randint(0, 3, (80,6))
marks[choice == 0] = torch.randint(1, 41, ((choice == 0).sum(),))
marks[choice == 1] = torch.randint(41, 71, ((choice == 1).sum(),))
marks[choice == 2] = torch.randint(71, 96, ((choice == 2).sum(),))
```

```python
marks = marks.float()
marks
```

```python
train = marks[:56]
valid = marks[56:68]
test = marks[68:80]

train.shape, valid.shape, test.shape
```

```python
X_train = train[:, :5]
y_train = train[:, 5].view(-1,1)
X_valid = valid[:, :5]
y_valid = valid[:, 5].view(-1,1)
X_test = test[:, :5]
y_test = test[:, 5].view(-1,1)
mean = X_train.mean(dim = 0)
std = X_train.std(dim = 0)
X_train = (X_train - mean) / std
X_valid = (X_valid - mean) / std
X_test = (X_test - mean) / std
X_train.shape, y_train.shape, X_valid.shape, y_valid.shape, X_test.shape, y_test.shape
```

## Creating a nn from scratch

```python
torch.manual_seed(42)
```

input(5) -->relu---> hidden(8)---> output(1)

```python
input_size = 5
hidden_size = 8
output_size = 1

W1 = torch.randn(input_size, hidden_size) *0.01
b1 = torch.zeros(hidden_size)
W2 = torch.randn(hidden_size, output_size) *0.01
b2 = torch.zeros(output_size)
W1.shape, b1.shape, W2.shape, b2.shape
```

```python
def relu(x):
  return torch.maximum(torch.zeros_like(x), x)
```

```python
def relu_derivative(x):
    grad = torch.zeros_like(x)
    grad[x > 0] = 1
    return grad
```

```python
def forward(x):
  z1 = x @ W1 + b1
  a1 = relu(z1)
  z2 = a1 @ W2 + b2
  return z1, a1, z2
```

```python
def mse_loss(y_pred, y):

    loss = ((y_pred - y)**2).mean()

    return loss
```

```python
lr = 0.01
epochs = 500
N = X_train.shape[0]

for epoch in range(epochs):

  # forward
  z1,a1, y_pred = forward(X_train)
  loss = mse_loss(y_pred, y_train)

  # backward
  dZ2 = (2/N)*(y_pred-y_train)
  dW2 = a1.T @ dZ2
  db2 = dZ2.sum(dim=0)

  dA1 = dZ2 @ W2.T
  dZ1 = dA1 * relu_derivative(z1)

  dW1 = X_train.T @ dZ1
  db1 = dZ1.sum(dim=0)

  # ===== UPDATE =====

  W1 -= lr * dW1
  b1 -= lr * db1

  W2 -= lr * dW2
  b2 -= lr * db2

  if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss {loss.item():.4f}")
```

```python
with torch.no_grad():

    _, _, val_pred = forward(X_valid)
    val_loss = mse_loss(val_pred, y_valid)

print("Validation Loss:", val_loss)
```

## creating nn in pytorch

```python
import torch.nn as nn
class SimpleNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(5,8)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(8,1)

  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
```

```python
model = SimpleNN()
print(model)
```

```python
sum(p.numel() for p in model.parameters())
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
epochs = 10000

for epoch in range(epochs):
  model.train()
  y_pred = model(X_train)
  loss = criterion(y_pred, y_train)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch%1000==0:
    print(f"Epoch {epoch}, Loss {loss.item():.4f}")
```

```python
model.eval()

with torch.no_grad():

    val_pred = model(X_valid)
    val_loss = criterion(val_pred, y_valid)

print("Validation Loss:", val_loss.item())
```

```python
with torch.no_grad():

    test_pred = model(X_test)
    test_loss = criterion(test_pred, y_test)

print("Test Loss:", test_loss.item())
```

### creating a deeper nn

```python
y_mean = y_train.mean()
y_std = y_train.std()

y_train = (y_train - y_mean) / y_std
y_valid = (y_valid - y_mean) / y_std
y_test  = (y_test - y_mean) / y_std
```

```python
X_train = X_train + 0.01 * torch.randn_like(X_train)
```

```python
class BiggerNN(nn.Module):

    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.Linear(5,32),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(32,16),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(16,1)
        )
    def forward(self,x):
        return self.net(x)

model = BiggerNN()
print(model)
```

```python
sum(p.numel() for p in model.parameters())
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.01,
    weight_decay=1e-3
)
```

```python
best_val = float("inf")
patience = 200
counter = 0

for epoch in range(5000):

    model.train()
    pred = model(X_train)
    loss = criterion(pred, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    model.eval()
    with torch.no_grad():
        val_loss = criterion(model(X_valid), y_valid)

    if val_loss < best_val:
        best_val = val_loss
        counter = 0
    else:
        counter += 1

    if counter >= patience:
        print(f"Early stopping at epoch {epoch}")
        break
```

```python
model.eval()

with torch.no_grad():

    train_loss = criterion(model(X_train), y_train)
    val_loss = criterion(model(X_valid), y_valid)
    test_loss = criterion(model(X_test), y_test)

print("Train Loss:", train_loss.item())
print("Valid Loss:", val_loss.item())
print("Test Loss:", test_loss.item())
```

```python
import torch
import torch.nn as nn
import pytorch_lightning as L
from torch.utils.data import DataLoader, TensorDataset
from pytorch_lightning.callbacks import EarlyStopping
```

```python
# Dataset + Dataloader
# -----------------------------
train_ds = TensorDataset(X_train, y_train)
val_ds   = TensorDataset(X_valid, y_valid)
test_ds  = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_ds, batch_size=len(train_ds))
val_loader   = DataLoader(val_ds, batch_size=len(val_ds))
test_loader  = DataLoader(test_ds, batch_size=len(test_ds))
```

```python
class BiggerNNLightning(L.LightningModule):

    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.Linear(5,32),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(32,16),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(16,1)
        )

        self.criterion = nn.MSELoss()

    def forward(self,x):
        return self.net(x)

    def training_step(self, batch, batch_idx):
        x,y = batch
        pred = self(x)
        loss = self.criterion(pred,y)

        self.log("train_loss", loss, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x,y = batch
        pred = self(x)
        loss = self.criterion(pred,y)

        self.log("val_loss", loss, prog_bar=True)

    def test_step(self, batch, batch_idx):
        x,y = batch
        pred = self(x)
        loss = self.criterion(pred,y)

        self.log("test_loss", loss)

    def configure_optimizers(self):
        return torch.optim.Adam(
            self.parameters(),
            lr=0.01,
            weight_decay=1e-3
        )
```

```python
model = BiggerNNLightning()

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=200,
    mode="min"
)

trainer = L.Trainer(
    max_epochs=5000,
    callbacks=[early_stop],
    log_every_n_steps=1
)
```

```python
trainer.fit(model, train_loader, val_loader)
```

```python
trainer.test(model, test_loader)
```
